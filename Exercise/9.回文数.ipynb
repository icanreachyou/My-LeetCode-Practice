## 9.3我的代码

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        temp = True
        if x >= -2**31 and x <=2**31-1:
            if x < 0:
                temp = False
            else:
                x_ch = x
                count = 0
                # 记录位数
                while x_ch >= 1:
                    x_ch = x_ch/10
                    count = count+1
                x_ch = x
                while count-2 >= 0 :
                    if int(x_ch/(10**(count-1))) != int(x_ch%10):
                        temp = False
                        break
                    x_ch = x_ch%(10**(count-1))
                    x_ch = x_ch/10
                    count = count-2
        return temp
```

## 9.4官方代码

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        return str(x)[ : : -1] == str(x)

# 作者：in3tantc
# 链接：https://leetcode-cn.com/problems/palindrome-number/solution/python3-li-yong-qie-pian-jie-jue-hui-wen-pnxq/
# 来源：力扣（LeetCode）
# 著作权归作者所有。商业转载请联系作者获得授权，非商业转载请注明出处。
```
