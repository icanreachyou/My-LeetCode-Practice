## 6.3我的代码

```python
class Solution:
    def convert(self, s: str, numRows: int) -> str:
        temp = ['']*numRows
        if numRows==1:
            return s
        else:
            for i in range(len(s)):
                k = i%(2*numRows-2)
                if k<= numRows-1:
                    temp[k] = temp[k] + s[i]
                else:
                    temp[2*(numRows-1)-k] = temp[2*(numRows-1)-k] + s[i]
            result = ''.join(temp)
            return result
```
